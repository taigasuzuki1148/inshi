```python
n =10000
```

```python
matrix = np.zeros((n,n), dtype=np.complex)

for k in range(n):
    matrix[k][k] = 2
    if k < n-1:
        matrix[k][k+1] = -1
        matrix[k+1][k] = -1

matrix_2 = np.zeros((n,n), dtype=np.complex)

for k in range(n):
    matrix_2[k][k] = 2
    if k < n-1:
        matrix_2[k][k+1] = -1
        matrix_2[k+1][k] = -1
    matrix_2[0][n-1] =-1
    matrix_2[n-1][0] =-1
```
