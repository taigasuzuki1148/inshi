```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns
from scipy.stats import binom

def func(x):
    return (0.3+0.7*np.exp(-x*0.4))/(1+np.exp(-x*0.4))

# グラフにプロット
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
x = np.linspace(-20,20,1000)
ax.tick_params(labelsize=12)
ax.plot(x, func(x),color="black")
ax.set_xlabel(r'$\beta$',fontsize=12)
ax.set_ylabel(r'$E$',fontsize=12)
ax.text(14, 0.33, r'$E \to \varepsilon_0$',fontsize=15)
ax.text(-14,0.65, r'$E \to \varepsilon_1$',fontsize=15)
ax.text(1,0.5, r'$E = \frac{\varepsilon_0+\varepsilon_1}{2}$',fontsize=15)
ax.legend()
ax.grid()
```

```python
fig.savefig("row_graph.pdf")
```

```python
def func(E):
    beta = -1/(0.4)*np.log((E-0.3)/(0.7-E))
    S = beta*0.4*1/(1+np.exp(beta*0.4)) + np.log(1+np.exp(-beta*0.4))
    return S 

# グラフにプロット
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
x = np.linspace(-0.2,0.6,1000)
ax.tick_params(labelsize=12)
ax.plot(x, func(x),color="black")
ax.set_xlabel(r'$E$',fontsize=12)
ax.set_ylabel(r'$S/k_B$',fontsize=12)
ax.legend()
ax.grid()
```
